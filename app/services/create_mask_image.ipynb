```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像を読み込む（カラー画像を想定）
image = cv2.imread("../tmp/Input_Image_No_1.png")

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = cv2.imread("../tmp/Input_Image_No_1.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(8, 8))
plt.title("gray_image")
# plt.imshow(image, cmap='gray')
plt.imshow(gray_image)
plt.show()


_, binary_image_140 = cv2.threshold(gray_image, 140, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(8, 8))
plt.title("140_binary_image")
# plt.imshow(image, cmap='gray')
plt.imshow(binary_image_140)
plt.show()


# 二値化処理を行う
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(8, 8))
plt.title("128_binary_image")
# plt.imshow(image, cmap='gray')
plt.imshow(binary_image)
plt.show()

smooth_image = cv2.blur(gray_image, (5, 5))
median_filter_image = cv2.medianBlur(gray_image, 9)

plt.figure(figsize=(8, 8))
plt.title("smooth_image")
plt.imshow(smooth_image)
plt.show()

plt.figure(figsize=(8, 8))
plt.title("median_filter_image")
plt.imshow(median_filter_image)
plt.show()

kernel = np.ones((3, 3), np.uint8)
dilated_image = cv2.dilate(binary_image_140, kernel, iterations=1)
eroded_image = cv2.erode(dilated_image, kernel, iterations=1)

plt.figure(figsize=(8, 8))
plt.title("eroded_image")
plt.imshow(eroded_image)
plt.show()

edges = cv2.Canny(gray_image, threshold1=128, threshold2=255)

plt.figure(figsize=(8, 8))
plt.title("edges")
plt.imshow(edges)
plt.show()
```
