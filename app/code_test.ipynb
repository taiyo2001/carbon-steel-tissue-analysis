```python
# ネットワークの保存

CHECKPOINT_PATH = APP_PATH + '/data/model'
MODEL_FILE = '/model_sample_weight.pth'

checkpoint_path = 'model/chkpoint_'
best_model_path = 'model/bestmodel.pt'

device = 'cuda' if torch.cuda.is_available() else 'cpu'


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3, 3)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        return x

# model = Net().to(device)
model = Net()

checkpoint = {
    'state_dict': model.state_dict(),
}

checkpoint_file = "/checkpoint_{}_cell_weight.pth".format(1)

print('model.state_dict() type:{}'.format(type(model.state_dict())))
print('checkpoint type:{}'.format(type(checkpoint)))
print(CHECKPOINT_PATH + checkpoint_file)

torch.save(checkpoint, CHECKPOINT_PATH + checkpoint_file)
load_model = torch.load(CHECKPOINT_PATH + checkpoint_file)
print('load_model:{}'.format(load_model))
```
